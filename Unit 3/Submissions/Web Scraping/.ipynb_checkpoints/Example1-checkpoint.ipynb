```python
import requests
import csv
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
# Find Mia Sheridan books on abebooks webpage 
url = 'https://www.abebooks.com/servlet/SearchResults?sts=t&cm_sp=SearchF-_-home-_-Results&an=mia+sheridan&tn=&kn=&isbn='
```

```python
#query the website and return the html 
page = requests.get(url)
```

```python
#parse the html using the Beautiful Soup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#get the price
price_box = soup.find_all('div', attrs={'class':'srp-item-price'})

prices = []
for price in price_box:
    price = price.text.strip()
    price = float(price[4:])
    prices.append(price)
```

```python
#get the title
# first all h2 tags to find the part of the the webpages with titles
heds = soup.find_all('h2')

# In h2 the a links are nested and the a links contain the span with title of the book
links = []
for h in heds:
    a = h.find('a')
    links.append(a)

spans = []    
for l in links:
    span = l.find('span')
    span = span.text.strip()
    spans.append(span)

    
# creating the dictionary and then data frame from the titles and prices
import pandas as pd

dict1 = {"Book title": spans, "Price": prices}
df = pd.DataFrame(dict1)
```
